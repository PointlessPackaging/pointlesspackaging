```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# convenient display function
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
def show_sift(img_logo, img_box):
    '''
    Computer SIFT descriptors for the logo and the image of packaging.
    Finds the 50 closest descriptors and visualizes the relationship
    to highlist the presence of a logo on packaging.
    '''
    
    # Create sift object
    sift = cv2.xfeatures2d.SIFT_create()
    
    # Initiate ORB detector
    orb = cv2.ORB_create()

    # find the keypoints and descriptors with ORB
    kp1, des1 = orb.detectAndCompute(img_logo,None)
    kp2, des2 = orb.detectAndCompute(img_box,None)

    # create BFMatcher object
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

    # Match descriptors.
    matches = bf.match(des1,des2)

    # Sort them in the order of their distance.
    matches = sorted(matches, key = lambda x:x.distance)

    # Draw first 50 matches.
    matches_visual = cv2.drawMatches(img_logo,kp1,img_box,kp2,matches[:50],None,flags=2)
    
    # show the matches
    display(matches_visual)
```

# Using SIFT descriptors to associate a company with packaging.

## Good matches
These are matches that I think we could possibly use to identify the company from what's on the box.

```python
amazon_box = cv2.imread('boxes/amazon_1.jpg', 0)
amazon_logo = cv2.imread('logos/amazon_logo.jpg', 0)
show_sift(amazon_logo, amazon_box)
```

```python
amazon_box = cv2.imread('boxes/amazon_4.jpg', 0)
amazon_logo = cv2.imread('logos/amazon_logo.jpg', 0)
show_sift(amazon_logo, amazon_box)
```

## Inconclusive attempts

These are examples of when SIFT descriptors completely fail.

```python
target_box = cv2.imread('boxes/target_1.jpg', 0)
target_logo = cv2.imread('logos/target_logo.jpg', 0)
show_sift(target_logo, target_box)
```
